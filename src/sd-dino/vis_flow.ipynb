```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
RESOLUTION = 128
```

```python
path = '../output/flow_images/'
image_name = 'shoe-aqua_cyan_right_7__shoe-asifn_yellow_right_3.npy'
flow = np.load(path + image_name).astype(np.float32)
```

```python
hsv = np.zeros((RESOLUTION, RESOLUTION, 3), dtype=np.uint8)
hsv[..., 1] = 255

mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
hsv[..., 0] = (ang * 180 / np.pi / 2)
hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(rgb)
```
